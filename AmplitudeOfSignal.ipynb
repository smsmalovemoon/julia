```julia
using PyPlot
```

```julia
using Distributions
```

```julia
X=linspace(-2,2)
a=1
b=2
n0= 100/(a+b)
x0= 0
w= 5/2.35
k= length(X)
D=N= zeros(1,k)
D =n0 .* (b .+ a .* exp((-1.*(X .- x0).^2)./ 2*w^2))
for i = 1:k
    N[i]= rand(Poisson(D[i]))
end
A_range=B_range=[0:0.01:3...] 
A_grid= repmat(A_range',length(B_range),1)
B_grid = repmat(B_range,1,length(A_range)) 
timer=1
D=zeros(length(A_range)*length(B_range),k)
for i = 1: length(A_range), j=1:length(B_range)
    D[timer,:] =n0*(A_grid[i,j] * exp(-(X - x0).^2 / 2*w^2) + B_grid[i,j])
    timer=timer+1
end
L= sum((N .*log(D)- D),2) 
LMax= maximum(L)
NewL= L- LMax
prob = exp(NewL)
prob_grid= zeros(length(A_range),length(B_range))
prob_grid = reshape(prob, length(A_range), length(B_range))
A1= sum(prob_grid',a)
B1= sum(prob_grid',b) 
plotHist= step(X,N')
```

```julia
contour(A_grid, B_grid,prob_grid')
```

```julia
plot(A_range,A1')
plot(A_range+0.2,A1',"--")
```

```julia
plot(B_range,B1)
```
