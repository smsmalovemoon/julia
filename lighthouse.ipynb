```julia
using PyPlot
```

```julia
function Lighthouse(k)
alpha=1
beta=1
theta = rand(-π/2:π/2,1,k)
X= beta .* tan(theta) .+ alpha 
alpha_range = [-5:0.1:5...]
L= -sum(log(beta^2 .+ (X .- alpha_range).^2),2)
LMax= maximum(L)
NewL= L .- LMax
prob = exp(NewL)
plot(alpha_range, prob)
xlabel("Lighthouse Position ")
ylabel("prob")
end
```

```julia
Lighthouse(1)
```
