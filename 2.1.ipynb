```julia
using PyPlot
```

```julia
tosses = 11
h=[0:0.1:1]
prob=ones(tosses)
##ignorance data##
plot(h,prob) ###fig1##
```

```julia
##the coin flipped once and get up head##
prob=h
plot(h,prob) ###fig2##
```

```julia
##the coin flipped asecond time and again comes up head##
prob=h.^2
plot(h,prob) ###fig3##
```

```julia
prob=h.^2.*(1-h)
plot(h,prob) ###fig4##
```

```julia
prob=h.^2.*(1-h).^2
plot(h,prob) ###fig5##
```

```julia
prob=h.^4.*(1-h).^6
plot(h,prob) ###fig6##
```

```julia
prob=h.^6.*(1-h).^12
plot(h,prob) ###fig7##
```

```julia
prob=h.^9.*(1-h).^24
plot(h,prob) ###fig8##
```

```julia
prob=h.^30.*(1-h).^70
plot(h,prob) ###fig9##
```

```julia
prob=h.^180.*(1-h).^320
plot(h,prob) ###fig10##
```

```julia
prob=h.^310.*(1-h).^690
plot(h,prob) ###fig11##
```
