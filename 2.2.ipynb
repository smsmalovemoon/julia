```julia
CoinFlip(H)=(rand()<0.25)?1:0
h= [0:0.001:1...]
N=0
Tosess= zeros(Int,1,N)
    for i=1:N
    Tosess[i]= CoinFlip(H)
    end    
prob= zeros(3,length(h))
prob_grid= zeros(3,2)
R= sum(Tosess)
likelihood= h.^R .* (1-h).^(N-R)
prob[1,:]= ones(Int,1,length(h))
for j=1:length(h)
    prob[2,j]= (exp(-(h[j]-0.5)^2 / (2*0.5^2))/(0.5*(2* pi)^(0.5)))
    prob[3,j]= (h[j])^(0.5-1)*(1-h[j])^(0.5-1)
    end
prob= likelihood' .* prob
NewP=prob
norm= [minimum(prob,2) maximum(prob,2)]
   
        if (norm[3,2] == Inf)
            c= find(prob[3,:] .== Inf)
           prob[3,c] = 1
            norm[3,2]= 1
            Pmax= maximum(prob[3,:])
            prob[3,c] = Pmax + 1
            norm[3,2]= Pmax+ 1
        end

        for j=2:3
    NewP[j,:] = (prob[j,:] - norm[j,1])./(norm[j,2] - norm[j,1])
        end 
plot(h,NewP')
```
